## K nearest neighbors model
Модель "К ближайших соседей"

В каком именно смысле "ближайших"?

как выбрать К?

```python
model = KNeighborsClassifier(n_neighbors=10)
```

```python
model.fit(x, y)
```

```python
x_test = np.linspace(-10, 10, 100)[:,np.newaxis]
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
f = plt.figure(figsize=(4,3), dpi=300)
p = plt.subplot(3,1,1)
_ = plt.hist(x[y==0], bins=20, color='red')
_ = plt.xlim(-10, 10)
_ = plt.xticks([])
_ = plt.ylabel('$y=0$')

p = plt.subplot(3,1,2)
_ = plt.hist(x[y==1], bins=20, color='blue')
_ = plt.xlim(-10, 10)
_ = plt.xticks([])
_ = plt.ylabel('$y=1$')

p = plt.subplot(3,1,3)
colors = {0:'red', 1:'blue'}
plt.scatter(x_test, y_pred, c=[colors[v] for v in y_pred], s=2)
_ = plt.xlim(-10, 10)
_ = plt.ylim(-0.2, 1.2)
_ = plt.xlabel('$x$')
_ = plt.ylabel('$y_{pred}$')

_ = f.patch.set_facecolor('white')
```
