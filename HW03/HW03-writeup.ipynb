# Задача восстановления регрессии

Вес = 0,6<br />
Срок сдачи - **01.12.2022**<br />
Решить задачу восстановления регрессии с применением модели линейной регресии. Допустимо порождение признаков (полиномиальных, экспоненциальных, логарифмических, etc.)<br />
Решение сдается оформленным в виде jupyter-notebook -файла `*.ipynb`<br />
В решении ожидается наличие следующих ключевых составляющих:<br />

- формулировка задачи, описание признаков объектов, формулировка функции ошибки, формулировка меры качества модели;
- исследование исходных данных на предмет скоррелированности признаков; фильтрация признаков (при необходимости); порождение признаков (при необходимости);
- выделение тестовой выборки для финальной оценки качества модели;
- оценка параметров модели линейной регрессии (решение задачи);
- аппроксмация целевой переменной на тестовой выборке;
- оценка качества полученной модели на тренировочной выборке, на тестовой выборке; сравнение этих показателей и вывод о переобучении или недообучении модели;
- аппроксимация целевой переменной на выборке вновь поступающих объектов (применение модели);

Визуализация:
- распределение признаков (гистограммы)
- распределение целевой переменной (гистограмма)
- графики зависимости целевой переменной от признаков на этапе применения модели.

Реализация линейной регрессии может быть написана с использованием библиотеки Numpy.<br />
Для порождения полиномиальных признаков можно использовать библиотеку scikit-learn.<br /><br />

Признаковое описание объектов обучающей выборки - в файле `X_train.npy`
